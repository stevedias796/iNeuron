5. What are the six different types of comparison operators? 
Answer: equal to (==), Greater than (>), Less than (<), Greater than equal to (>=), Less than equal to (<=), Not equal to (!=)


6. How do you tell the difference between the equal to and assignment operators? 
Answer: equal to operator is a comparison operator, we use it to check if two values are equal or no. assignment operator is used to assign a value to a variable. Also equal to operator is denoted as '==' whereas assignment is '='

11. How can you tell the difference between break and continue? 
Answer: break keyword is used when you want to end the loop or come out of the loop. continue keyword is used when you want to jump to next iteration in the loop.
